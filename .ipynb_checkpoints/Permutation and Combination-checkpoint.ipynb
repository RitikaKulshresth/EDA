```python
import math
```

```python
# 5!
math.factorial(5)
```

```python
# Permutation without repetition : nPr = n!/(n-r)!
math.perm(5,3)
```

```python
# Permutation with repetition : n^r
# n = 5, r = 4
math.pow(5,4)
```

```python
# Combination without repetition : nCr = n!/(r!*(n-r)!)
# n = 5 and r = 3
math.comb(5,3)
```

```python
# Combination with repetition : (n+r-1)Cr = (n+r-1)!/(r!*(n-1)!)
# n = 5
# r = 3
math.factorial(5+3-1)/(math.factorial(3)*math.factorial(5-1))
```

```python
# BANANA
# n = 6
# A = 3
# N = 2

math.factorial(6)/(math.factorial(3)*math.factorial(2))
```

```python
# PDF - Parametric Density Estimation
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import normal

sample = normal(loc=50, scale=5, size=1000)
```

```python
print(sample.mean())
print(sample.std())
```

```python
sample.min()
```

```python
sample.max()
```

```python
plt.hist(sample, bins=10)
```

```python
from scipy.stats import norm
dist = norm(49.8,5) # norm(mean, std)
```

```python
dist
```

```python
values = np.linspace(sample.min(), sample.max(), 100)
```

```python
values
```

```python
probabilities = [dist.pdf(value) for value in values]
```

```python
plt.hist(sample, bins=10, density = True)
plt.plot(values, probabilities)
```

```python
import seaborn as sns
sns.distplot(sample)
```

# KDE

```python
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size = 700)
sample = np.hstack((sample1, sample2))
```

```python
sample
```

```python
plt.hist(sample, bins=50)
```

```python
from sklearn.neighbors import KernelDensity
model = KernelDensity(bandwidth =5, kernel='gaussian')
sample = sample.reshape((len(sample),1))
model.fit(sample)
```

```python
sample.shape
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values),1))
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
```

```python
plt.hist(sample, bins=50, density = True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
sns.kdeplot(sample.reshape(1000), bw_adjust=1)
```
