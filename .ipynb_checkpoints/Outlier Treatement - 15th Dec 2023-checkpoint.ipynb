# Distribution Method - Z_Score concept

```python
dataset['cgpa'].skew()
```

```python
dataset['placement_exam_marks'].skew()
```

```python
dataset.describe()
```

```python
# highest z_score = mean + 3*std
# lowest z_score = mean - 3*std
print("Highest value", dataset['cgpa'].mean() + 3*dataset['cgpa'].std())
print("lowest value", dataset['cgpa'].mean() - 3*dataset['cgpa'].std())
```

```python
# Finding outliers
dataset[(dataset['cgpa'] > 8.80) | (dataset['cgpa'] < 5.11)]
```

# Approach no - 1 : Trimming

```python
new_df = dataset[(dataset['cgpa'] < 8.80) & (dataset['cgpa'] > 5.11)]
new_df
```

# Approach no - 2 : Capping method (very popular)

```python
dataset['cgpa_zscore'] = (dataset['cgpa'] - dataset['cgpa'].mean())/dataset['cgpa'].std()
```

```python
dataset
```

```python
dataset[(dataset['cgpa_zscore']>3) | (dataset['cgpa_zscore']< -3)]
```

```python
# Capping 
upper_limit = dataset['cgpa'].mean() + 3*dataset['cgpa'].std()
lower_limit = dataset['cgpa'].mean() - 3*dataset['cgpa'].std()
print("upper limit : ", upper_limit)
print("lower limit :", lower_limit)
```

```python
dataset['cgpa'] = np.where(dataset['cgpa'] > upper_limit, upper_limit,
                          np.where(dataset['cgpa'] < lower_limit, lower_limit,
                                 dataset['cgpa']  ))
```

```python
dataset['cgpa'].shape
```

```python
dataset['cgpa'].describe()
```

# Box plot method - InterQuartile concept

```python
dataset.columns
```

```python
dataset['placement_exam_marks'].describe()
```

```python
sns.boxplot(dataset['placement_exam_marks'])
```

```python
# Finding the IQR
Q1 = dataset['placement_exam_marks'].quantile(0.25)
Q3 = dataset['placement_exam_marks'].quantile(0.75)
IQR = Q3 - Q1

print("Percentile25 :", Q1)
print("Percentile75 :", Q3)
print("InterQuartileRange :", IQR)
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

print("Upper Limit :", upper_limit)
print("Lower Limit :", lower_limit)
```

```python
# Finding outlier
dataset[(dataset['placement_exam_marks'] > upper_limit) | (dataset['placement_exam_marks'] < lower_limit)]
```

```python
dataset[dataset['placement_exam_marks'] < lower_limit]
```

# Approach no 1 : Trimming approach

```python
new_df = dataset[dataset['placement_exam_marks'] < upper_limit]
```

```python
new_df.shape
```

```python
new_df['placement_exam_marks'].describe()
```

# Approach no - 2 : Capping method

```python
new_df_cap = dataset.copy()
```

```python
new_df_cap.head()
```

```python
new_df_cap['placement_exam_marks'] = np.where(new_df_cap['placement_exam_marks'] > upper_limit,
                                             upper_limit,
                                              np.where(new_df_cap['placement_exam_marks'] < lower_limit,
                                                       lower_limit, 
                                                      new_df_cap['placement_exam_marks']))
```

```python
new_df_cap.shape
```

```python
new_df_cap['placement_exam_marks'].describe()
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(dataset['placement_exam_marks'])
plt.subplot(2,2,2)
sns.boxplot(dataset['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['placement_exam_marks'])
plt.subplot(2,2,4)
sns.boxplot(new_df_cap['placement_exam_marks'])

plt.show()
```
