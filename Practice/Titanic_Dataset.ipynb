# Univariate Analysis

Categorical Variables: Use countplot to see the distribution.
    
Numerical Variables: Use histograms and box plots to observe distributions and outliers.

```python
final_df.head()
```

```python
# Categorical - Pclass,Name,Sex,Ticket,Embarked

sns.countplot(x='Pclass',data=final_df)
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(x='Sex',data=final_df)
plt.title('Sex')
plt.show()

sns.countplot(x='Embarked',data=final_df)
plt.title('Embarked')
plt.show()

# sns.countplot(x='Survived',data=final_df)
# plt.title('Survived')
# plt.show()
```

```python
# Numerical: Age, Fare
# Age	SibSp	Parch	Fare
final_df['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

final_df['SibSp'].hist(bins=30,edgecolor='black')
plt.title('Siblings Count')
plt.xlabel('SibSp')
plt.ylabel('Frequency')
plt.show()

final_df['Parch'].hist(bins=30,edgecolor='black')
plt.title('Parents count')
plt.xlabel('Parch')
plt.ylabel('Frequency')
plt.show()

final_df['Fare'].hist(bins=30,edgecolor='black')
plt.title('Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
